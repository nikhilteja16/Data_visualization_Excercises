Categorical plotting

```python
import seaborn as sns
tips = sns.load_dataset('tips')
```

```python
import numpy as np
import matplotlib.pyplot as plt
tips.head()
```

```python
sns.histplot(data = tips, x= 'tip', bins = 50)
plt.title('Counts of tip given by customers')
plt.xlabel('Tip Amount')
plt.ylabel('Count')
```

```python
sns.barplot(x='sex' ,y ='tip' , data = tips)
plt.title('Mean of the tips given by men and female')
```

```python
tips.groupby('sex')['tip'].mean()
```

```python
sns.barplot(x = 'sex', y = 'tip', data = tips, estimator= np.var)
plt.title('variance of tips by gender')
```

```python
sns.countplot(data = tips, x = 'sex')
plt.title('Count of number of people giving tips')
```

Boxplots are used to show the distribution of categorical data.


```python
sns.boxplot(data = tips, y = 'tip', hue = 'sex')
```

```python
tips
```

```python
sns.boxplot(data = tips , y= 'total_bill', x = 'day', hue = 'smoker' )
```

```python
 tips.groupby('day')['total_bill'].value_counts(ascending = False)
```

```python
sns.violinplot(x ='day', y = 'total_bill', data = tips, hue = 'sex', split = True)
```

```python
sns.stripplot(data = tips, x = 'day', y = 'total_bill', palette= 'Dark2', jitter= True, hue= 'smoker')
```

```python
tips.count()
```

Swarm plot is combining both a violin and a split plot

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = tips, palette= 'Dark2', hue = 'sex')
```

```python
sns.catplot(data = tips, x  = 'day', y = 'total_bill', kind  = 'box')
```

Matrix plots

```python
import seaborn as sns
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights.shape
```

```python
tips
```

```python
tips_corr = tips.corr()
```

```python
sns.heatmap(tips_corr,annot= True, cmap = 'coolwarm' )
```

```python
flights
```

```python

flights.pivot_table(index= 'year', columns = ['month'] ).plot(kind = 'box', )
plt.figure(figsize=(24,6))
plt.show()
```

```python
fl = flights.pivot_table(index = 'month', columns ='year' )
```

```python
sns.heatmap(fl, linecolor='white', cmap= 'coolwarm', linewidths = '0.35')
plt.figure(figsize=(12,6))
plt.show()
```

```python
g = sns.FacetGrid(data = tips, col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```
