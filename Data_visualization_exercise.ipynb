```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.shape
```

```python
sns.set_style('whitegrid')
```

```python
sns.displot(titanic['fare'], kde= False, bins = 30, color= 'red')
```

```python
titanic
```

```python
plt.figure(figsize= (6,6))
sns.boxplot(data = titanic, x = 'class', y = 'age')
```

```python
sns.swarmplot(data = titanic, x = 'class', y = 'age', palette= 'plasma')
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data = titanic,x = 'sex')
```

```python
titanic_corr = titanic.corr()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(titanic_corr, cmap = 'coolwarm', )
```

```python
titanic
```

```python
g = sns.FacetGrid(data = titanic, col = 'sex')
g.map(sns.histplot,data = titanic, x = 'age')
```

```python
sns.jointplot(data = titanic, x = 'fare', y = 'age', palette= 'inferno', color= 'green')
```
